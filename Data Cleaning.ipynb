Secondly, we can think about our next null value: AGE

```python
df.Age.isna().sum()
```

We have nearly 20 percent of our data missing here!

This is problematic because we can't just use the technique that we employed previously where we dropped the two embarked rows with nulls. And age can theoretically be very important to figuring out one's likelhood of survival. 

Here we will examine a variety of techniques we can use to impute or replace the values of Age in a way that let's us work onward without losing a large amount of potentially very useful data. 


```python
mean_fill = df.Age.fillna(np.mean(df.Age))
mean_fill
```

```python
df_mean_fill = df
df_mean_fill.Age=mean_fill
df_mean_fill.isna().sum()
```

Alternatively, we can use bfill and ffill to fill it in a different way. 

```python
f_fill = df.Age.fillna(method='ffill')
f_fill.head()
```

```python
df_ffill = df
df_ffill.Age=f_fill
df_ffill.isna().sum()
```
